```python
import pandas as pd
import numpy as np
import datetime
from time import strftime
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
base_data = pd.read_csv('Medical_Appointment_Data.csv')
```

```python
#modifying the date and time into standard form
base_data['ScheduledDay'] = pd.to_datetime(base_data['ScheduledDay']).dt.date.astype('datetime64[ns]')
base_data['AppointmentDay'] = pd.to_datetime(base_data['AppointmentDay']).dt.date.astype('datetime64[ns]')
```

```python
base_data.head()
```

For the schedule day and appointment day storing the weekdays only into a variable

```python
# 5 is Saturday, 6 is Sunday 

base_data['sch_weekday'] = base_data['ScheduledDay'].dt.dayofweek
base_data['app_weekday'] = base_data['AppointmentDay'].dt.dayofweek
```

```python
# Count number of scheduled appointments for each weekday

sch_appointments_count = base_data['sch_weekday'].value_counts().sort_index()
```

```python
# Create bar graph
plt.figure(figsize=(10, 6))
sch_appointments_count.plot(kind='bar', color='skyblue')
plt.xlabel('Weekday (0=Monday, 1=Tuesday, ..., 6=Sunday)')
plt.ylabel('Number of Appointments')
plt.title('Appointments per Scheduling Day Weekday')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count number of appointments for each weekday

appointments_count = base_data['app_weekday'].value_counts().sort_index()
```

```python
# Create bar graph
plt.figure(figsize=(10, 6))
appointments_count.plot(kind='bar', color='skyblue')
plt.xlabel('Weekday (0=Monday, 1=Tuesday, ..., 6=Sunday)')
plt.ylabel('Number of Appointments')
plt.title('Appointments per Scheduling Day Weekday')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot scheduling weekday appointments
plt.bar(sch_appointments_count.index, sch_appointments_count.values, color='skyblue', label='Scheduled Weekday')

# Plot appointment weekday appointments
plt.bar(appointments_count.index + 0.4, appointments_count.values, color='orange', label='Appointment Weekday')

plt.xlabel('Weekday (0=Monday, 1=Tuesday, ..., 6=Sunday)')
plt.ylabel('Number of Appointments')
plt.title('Comparison of Scheduling Weekday and Appointment Weekday')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
